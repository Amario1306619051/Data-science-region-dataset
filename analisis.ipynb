### Menghitung Jumlah Data yang Hilang dalam Setiap Kolom

```python
# Menghitung jumlah data yang hilang dalam setiap kolom
missing_data = dataset.isna().sum()

# Menampilkan hasil
print(missing_data)
```

Deskripsi Singkat:
Kode ini digunakan untuk menghitung jumlah data yang hilang (missing data) dalam setiap kolom (atribut) dari suatu dataset. Langkah-langkahnya adalah sebagai berikut:

1. `missing_data = dataset.isna().sum()`: Pada baris ini, kita menggunakan `.isna()` untuk mengidentifikasi nilai-nilai yang hilang dalam dataset. Hasilnya adalah DataFrame baru yang berisi `True` untuk setiap sel yang memiliki data yang hilang, dan `False` untuk yang lainnya. Kemudian, kita menggunakan `.sum()` pada DataFrame ini untuk menghitung jumlah nilai `True` dalam setiap kolom, sehingga kita mendapatkan jumlah data yang hilang dalam setiap kolom.

2. `print(missing_data)`: Baris ini digunakan untuk mencetak hasil perhitungan jumlah data yang hilang dalam setiap kolom ke layar atau konsol.

Dengan kode ini, Anda akan mendapatkan output berupa daftar atau Series yang menunjukkan jumlah data yang hilang dalam setiap kolom dari dataset yang digunakan. Informasi ini penting untuk menganalisis sejauh mana data yang hilang dalam dataset dan untuk memutuskan langkah-langkah selanjutnya dalam pengolahan data atau pemenuhan data yang hilang.
```

````python
# Menghitung jumlah masing-masing kategori
gender_counts = dataset['gender'].value_counts()

# Membuat pie chart
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Distribusi Gender')
plt.axis('equal')  # Agar diagram lingkaran terlihat bulat
plt.show()
````

````python
# Membaca dataset (misalnya, dataset disimpan dalam variabel df)
age_range = dataset['age_range']

# Membuat plot diagram batang
sns.countplot(data=dataset, x='age_range')
plt.title('Distribusi Age Range')
plt.xticks(rotation=45)  # Untuk mengatur label sumbu x
plt.show()
````

````python
#plot goup city, dan lokasi
grouped_data = dataset.groupby(['province', 'city'])
result = grouped_data.size().reset_index(name='count')
result
````

````python
# Menghitung jumlah masing-masing kota dalam setiap provinsi
city_counts = dataset.groupby(['province', 'city']).size().unstack().fillna(0)

for location in city_counts.index:
    plt.figure(figsize=(6, 6))
    pie_data = city_counts.loc[location]
    top_cities = pie_data.nlargest(5)  # Menampilkan hanya 5 kota dengan frekuensi tertinggi
    other_cities = pie_data.sum() - top_cities.sum()
    pie_data = top_cities.append(pd.Series([other_cities], index=['Others']))

    # Membuat pie chart
    plt.pie(pie_data, autopct='%1.1f%%', startangle=140)
    plt.title(f'Distribusi Kota di {location}')
    plt.axis('equal')  # Agar diagram lingkaran terlihat bulat

    # Menambahkan legend di luar pie chart
    plt.legend(title='Kota', labels=pie_data.index, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    plt.show()
````

````python
# Filter data untuk Jawa Barat
jabar_df = dataset1[dataset1['province'] == 'Jawa Barat']

# Menghitung distribusi gender di Jawa Barat
gender_counts = jabar_df['gender'].value_counts()

# Membuat pie chart untuk distribusi gender
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribusi Gender di Jawa Barat')
plt.axis('equal')
plt.show()

# Menghitung distribusi Age Range di Jawa Barat
age_range_counts = jabar_df['age_range'].value_counts() 

# Membuat pie chart untuk distribusi Age Range
plt.figure(figsize=(8, 8))
_, _, autotexts = plt.pie(age_range_counts, labels=age_range_counts.index, autopct='%1.1f%%', startangle=140)

# Menambahkan legend dengan label yang sesuai
plt.title('Distribusi Age Range di Jawa Barat')
plt.axis('equal')

# Menampilkan legend di luar pie chart
legend_labels = age_range_counts.index
plt.legend(legend_labels, title='Age Range', loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Menambahkan label persentase di sekitar pie chart
for autotext in autotexts:
    autotext.set_size(12)
    autotext.set_color('white')

plt.show()


# Visualisasi Interest di Jawa Barat (contoh word cloud)

interest_text_jabar = ' '.join(jabar_df['interest'])
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(interest_text_jabar)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Minat Umum di Jawa Barat')
plt.axis('off')
plt.show()
````

````python
#mengubah category

#memanfaatkan fungsi deh
def categorize_age(age):
    if 10 <= age < 25:
        return 'Gen Z'
    elif 25 <= age < 40:
        return 'Gen Millennial'
    elif 40 <= age < 60:
        return 'Gen X'
    elif age >= 60:
        return 'Baby Boomers'
    elif age < 10:
        return 'Gen Alpha'
    else:
        return 'Unknown'


#menambahkan semua kolom
dataset['age_category'] = dataset['age'].apply(categorize_age)
age_category_counts = dataset['age_category'].value_counts()
````
